```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import time
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
class GDRegresson:
    def __init__(self,learning_rate=0.01,epochs=100):
        self_cofe_ = None
        self_intercep_ = None
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,X_train,y_train):
        # init your coefs
        self.intercep_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            # update all the coef and the intercept
            y_hat = np.dot(X_train,self.coef_) + self.intercep_
            # print("shape of y_hat",y_hat.shape)
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercep_ = self.intercep_ - (self.lr * intercept_der)

            coef_der = -2 *np.dot((y_train - y_hat),X_train)/X_train.shape[0]
            self.coef_ = self.coef_ - (self.lr * coef_der)
            
        print(self.intercep_, self.coef_)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercep_
```

```python
gdr = GDRegresson(epochs = 100,learning_rate = 0.01)
```

```python
start = time.time()
gdr.fit(X_train,y_train)
print("The time taken is",time.time() - start)
```

```python
y_pred = gdr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
